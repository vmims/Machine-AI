```python

# Import necessary libraries
from sklearn.cluster import KMeans

# Step 3.1: Filter employees who left and select relevant columns
left_employees = df[df['left'] == 1][['satisfaction_level', 'last_evaluation']]

# Step 3.2: Perform K-means clustering into 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
left_employees['Cluster'] = kmeans.fit_predict(left_employees)

# Display the first few rows of clustered data
left_employees.head()

# Visualize the clusters
plt.figure(figsize=(12, 6))
sns.scatterplot(data=left_employees, x='satisfaction_level', y='last_evaluation', hue='Cluster', palette='viridis')
plt.title('Clustering of Employees Who Left Based on Satisfaction and Evaluation')
plt.xlabel('Satisfaction Level')
plt.ylabel('Last Evaluation Score')
plt.legend(title='Cluster')
plt.show()
```

Cluster 0 suggests dissatisfaction and poor performance are strong indicators of leaving.
Cluster 1 shows that even happy, high-performing employees might leave, suggesting other retention strategies are needed for this group.
Cluster 2 indicates that even well-evaluated employees with mixed satisfaction levels need attention to ensure they stay

```python
cluster_summary = left_employees.groupby('Cluster').agg({
    'satisfaction_level': 'mean',
    'last_evaluation': 'mean'
})
print(cluster_summary)

# Detailed Description of Each Cluster
for cluster in range(3):
    print(f"\nCluster {cluster} Summary:")
    print(left_employees[left_employees['Cluster'] == cluster].describe())

# Visualize the clusters with added context
plt.figure(figsize=(12, 8))
sns.scatterplot(x='satisfaction_level', y='last_evaluation',
                hue='Cluster', data=left_employees, palette='viridis', s=100)
plt.title('Clustering of Employees Who Left Based on Satisfaction and Evaluation')
plt.xlabel('Satisfaction Level')
plt.ylabel('Last Evaluation Score')
plt.legend(title='Cluster')
plt.axvline(x=0.5, linestyle='--', color='red', label='Satisfaction Threshold')
plt.axhline(y=0.6, linestyle='--', color='blue', label='Evaluation Threshold')
plt.show()
```

Low Satisfaction, High Evaluation: These employees might be high performers but are unhappy, possibly due to stress or feeling overworked.
Low Satisfaction, Low Evaluation: This group could be disengaged or unhappy with their work, leading to poor performance and a higher chance of leaving.
High Satisfaction, High Evaluation: These employees are happy and perform well, but they may have left due to reasons outside of work, like better job offers or personal choices.



*  Train a Random Forest Classifier model, apply the 5-fold CV, and plot

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, auc, confusion_matrix # Import confusion_matrix instead of plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay # Import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Define the Random Forest model
rf_model = RandomForestClassifier(random_state=123)

# Define 5-Fold Cross-Validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

# Use cross_val_predict to get predictions across the folds
y_pred = cross_val_predict(rf_model, X_train, y_train, cv=cv, method='predict')

# Display classification report
print("Classification Report:\n")
print(classification_report(y_train, y_pred))

# Fit the model on the training data
rf_model.fit(X_train, y_train)

# Generate the confusion matrix
cm = confusion_matrix(y_test, rf_model.predict(X_test), labels=rf_model.classes_) # Generate confusion matrix using confusion_matrix()

# Plot the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Stayed", "Left"]) # Use ConfusionMatrixDisplay to create the plot
disp.plot(cmap=plt.cm.Blues) # Plot the confusion matrix with desired colormap
plt.title('Confusion Matrix - Random Forest')
plt.show()

# Optional: Plotting the ROC curve
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]  # Getting probabilities of the positive class
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Random Forest (AUC = {roc_auc:.2f})', color='blue')
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal reference line
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Random Forest')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming you have the following models already fitted
# logistic_model, rf_model, gb_model are your trained models
# X_test and y_test are your test data

# Function to plot ROC curve for each model
def plot_roc_curve(model, X_test, y_test, model_name, color):
    try:
        # Predict probabilities of the positive class
        y_pred_proba = model.predict_proba(X_test)[:, 1]

        # Compute ROC curve and AUC score
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)

        # Plot ROC curve
        plt.plot(fpr, tpr, color=color, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')
        print(f"{model_name} AUC Score: {roc_auc:.4f}")
    except Exception as e:
        print(f"Error in plotting ROC for {model_name}: {e}")

# Plotting the ROC curves
plt.figure(figsize=(10, 8))

# Plot ROC for Logistic Regression
plot_roc_curve(logistic_model, X_test, y_test, 'Logistic Regression', 'blue')

# Plot ROC for Random Forest
plot_roc_curve(rf_model, X_test, y_test, 'Random Forest', 'green')

# Plot ROC for Gradient Boosting
plot_roc_curve(gb_model, X_test, y_test, 'Gradient Boosting', 'red')

# Add diagonal line representing no discrimination
plt.plot([0, 1], [0, 1], 'k--', lw=2)

# Labels and title
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Function to plot confusion matrix
def plot_confusion_matrix(model, X_test, y_test, model_name):
    try:
        # Predict the class labels for the test set
        y_pred = model.predict(X_test)

        # Calculate the confusion matrix
        cm = confusion_matrix(y_test, y_pred)

        # Plot the confusion matrix
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Stayed', 'Left'])
        disp.plot(cmap=plt.cm.Blues)
        plt.title(f'Confusion Matrix for {model_name}')
        plt.grid(False)
        plt.show()

        # Print the confusion matrix
        print(f"Confusion Matrix for {model_name}:\n{cm}\n")
    except Exception as e:
        print(f"Error generating confusion matrix for {model_name}: {e}")

# Assuming logistic_model, rf_model, and gb_model are your trained models
# X_test and y_test are your test data

# Confusion matrix for Logistic Regression
plot_confusion_matrix(logistic_model, X_test, y_test, 'Logistic Regression')

# Confusion matrix for Random Forest
plot_confusion_matrix(rf_model, X_test, y_test, 'Random Forest')

# Confusion matrix for Gradient Boosting
plot_confusion_matrix(gb_model, X_test, y_test, 'Gradient Boosting')
```

When predicting which employees might leave, its more important to focus on Recall than Precision. Recall tells us how well the model finds employees who are likely to quit, ensuring we dont miss anyone who needs attention. Missing someone at risk (low Recall) means losing them without a chance to keep them. Although Precision shows how accurate the model is in predicting only those truly at risk, its less critical in this case because wed rather identify more at-risk employees, even if a few arent actually planning to leave. So, focusing on Recall helps catch most of those who might leave, allowing us to act and retain them.
